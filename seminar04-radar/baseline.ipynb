```python
import pandas as pd
import numpy as np
```

```python
def precission(y_hat, y_true):
    tp = np.sum(y_true[y_hat == 1])
    tot_p = np.sum(y_hat)
    
    return tp / tot_p

def recall(y_hat, y_true):
    tp = np.sum(y_true[y_hat == 1])
    fn = np.sum(y_true[y_hat==0])
    
    return tp / (tp + fn)
```
